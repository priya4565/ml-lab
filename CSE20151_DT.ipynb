```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Week_9/titanic.csv')
df
```

```python
df.describe().T
```

```python
df.isna().any
```

```python
df1=df.drop(columns='Cabin')
```

```python
df1
```

```python
df1.isna().any
```

```python
df1=df1.dropna()
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df1['Sex']=lb.fit_transform(df1['Sex'])
```

```python
df1
```

```python
ip=df1.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Embarked'],axis='columns')
op=df1['Survived']
```

```python
ip
```

```python
op
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier(criterion="entropy")
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(ip,op,test_size=0.2,random_state=42)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print('Model train accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(ytest, ypred)))
```

```python
ytpred=model.predict(xtrain)
```

```python
print('Model train accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(ytrain, ytpred)))
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(model.fit(xtrain,ytrain))

```
